```python
articles = []
for link in links:
    reqs = requests.get(link)
    soup = BeautifulSoup(reqs.text, 'lxml')
    content = soup.find(class_='cell lg-9 md-12 content-block')
    
    for a_tag in content.find_all('a'):
        href = a_tag.get('href', '')  # Get the link from the href attribute
        text = a_tag.text
        a_tag.replace_with(text + ":" + " " + href + " ")
    article = {}
    article['source'] = link
    article['content'] = content.get_text()
    articles.append(article)
    
```

# SPLIT DOCUMENTS INTO CHUNKS

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n", ".", " ", ""],
    chunk_size=1000,
    chunk_overlap=200
)
```

```python
from langchain.docstore.document import Document
```

```python
def get_article(article):
    return Document(
        page_content=article['content'],
        metadata={"source": article['source']}
    )
```

```python
sources = [get_article(article) for article in articles]    
```

```python
len(sources)
```

```python
sources
```

```python
source_chunks = []
for source in sources:
    for chunk in splitter.split_text(source.page_content):
        source_chunks.append(Document(page_content=chunk, metadata=source.metadata))
```

```python
len(source_chunks)
```

```python
source_chunks
```
